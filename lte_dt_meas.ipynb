### 3D Graph Visualizayion

Langkah pertama adalah melakukan pengolahan data. Data harus dikonversikan kedalam bentuk (format) .csv, atau bisa juga dalal format lain selama dapat dibaca oleh python, i.e format JSON.

Kemudian, data kosong, outlier, salah format, dll juga harus dibersihkan.

Kemudian, tampilkan dalam bentuk grafik.


```python
# import libraries
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import pandas as pd

# read and load data csv file into DataFrame (table)
lte_dt_meas = pd.read_csv('lte_dt_meas.csv')

# preparing x y z axis
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# get particular data to be presented in graph
x = lte_dt_meas['LTE_UE_RSRP'].values
y = lte_dt_meas['LTE_UE_SINR'].values
z = lte_dt_meas['LTE_UE_Power_Tx'].values

# scatter plot
ax.scatter(x, y, z, c='r', marker='o')
# ax.scatter(x, y, z, c='b', marker='*')    # color blue & marker "*"

# Title
# Placement 0, 0 would be the bottom left, 1, 1 would be the top right.
ax.text2D(0.35, 0.95, "LTE UE MEASUREMENT", transform=ax.transAxes)
ax.text2D(0.40, 0.85, "WEEK-05 2022", transform=ax.transAxes)

# axis label
ax.set_xlabel('RSRP')
ax.set_ylabel('SINR')
ax.set_zlabel('Power_Tx')

# show the graph
plt.show()
```

Kita juga dapat melihat korelasi datanya.
